```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python


# Corrected file path
file_path = r'C:\Users\ekomo\Downloads\Stores.csv'

# Read the CSV file
supermarket_store = pd.read_csv(file_path)

# Print the DataFrame
print(supermarket_store)
```

```python
# 1. Data Exploration
print(supermarket_store.info())
print(supermarket_store.describe())
print(supermarket_store.head())
```

```python
key_metrics = ['Store_Sales', 'Daily_Customer_Count', 'Items_Available', 'Store_Area']
```

```python
for metric in key_metrics:
    print(f"Descriptive statistics for {metric}:\n")
    print(supermarket_store[metric].describe())
```

```python
for metric in key_metrics:
    sns.histplot(supermarket_store[metric], kde=True)
    plt.title(f'Distribution of {metric}')
    plt.show()
```

```python
top_performing_stores = supermarket_store.nlargest(10, 'Store_Sales')
bottom_performing_stores = supermarket_store.nsmallest(10, 'Store_Sales')
print("Top-performing stores:")
print(top_performing_stores)
print("Bottom-performing stores:")
print(bottom_performing_stores)



```

```python
# 7. Correlation Analysis
correlation_matrix = supermarket_store[key_metrics].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu")
plt.title("Correlation Matrix")
plt.show()



```
